```python
fuel_types = data['Fuel_Type'].unique()

plt.figure(figsize=(10, 8))

for i, fuel in enumerate(fuel_types):
    plt.subplot(len(fuel_types), 1, i + 1)
    subset = data[data['Fuel_Type'] == fuel]
    plt.hist(subset['Price'], bins=10, color='skyblue', edgecolor='black')
    plt.title(f'Price Distribution - {fuel}')
    plt.xlabel('Price')
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
