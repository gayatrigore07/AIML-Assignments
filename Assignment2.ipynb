```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
```

```python
df_sal = pd.read_csv('C:/Users/goreg/Downloads/Salary_Data.csv')
df_sal.head()
```

```python
df_sal.describe()
```

```python
plt.scatter(df_sal['YearsExperience'], df_sal['Salary'], color = 'lightcoral')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.box(False)
plt.show()
```

```python
X = df_sal.iloc[:, :1] 
y = df_sal.iloc[:, 1:]
```

```python
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)     
y_pred_train = regressor.predict(X_train)
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```

```python
residuals_test = y_test - y_pred_test
```

```python
residuals_train = y_train - y_pred_train
```

```python
plt.scatter(X_train, residuals_train, color='teal')
plt.hlines(y=0, xmin=X_train.min(), xmax=X_train.max(), color='gray', linestyle='--')
plt.title('Residuals (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Residuals')
plt.box(False)
plt.show()
```

```python
mae=mean_absolute_error(y_test, y_pred_test)
mse=mean_squared_error(y_test, y_pred_test)
rmse=mean_squared_error(y_test, y_pred_test, squared=False) 
r2=r2_score(y_test, y_pred_test)

print(f"Mean Absolute Error:{mae:.2f}")
print(f"Mean Squared Error:{mse:.2f}")
print(f"Root Mean Squared Error:{rmse:.2f}")
print(f"R-squared:{r2:.2f}")
```
